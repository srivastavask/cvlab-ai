```python
#Q2.
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'color_image.jpeg'
color_image = cv2.imread(image_path)
color_image_rgb = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

#histograms
r_channel, g_channel, b_channel = cv2.split(color_image)

plt.subplot(1,2,1)
plt.imshow(color_image_rgb)
plt.title('Original Color Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray_image)
plt.title('Grayscale Image')
plt.axis('off')

# Mode 1: Histogram (Frequency of Occurrence)
r_hist = cv2.calcHist([r_channel], [0], None, [256], [0, 256])
g_hist = cv2.calcHist([g_channel], [0], None, [256], [0, 256])
b_hist = cv2.calcHist([b_channel], [0], None, [256], [0, 256])
gray_hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Mode 2: Probability of Occurrence
total_pixels = color_image.shape[0] * color_image.shape[1]

r_hist_prob = r_hist / total_pixels
g_hist_prob = g_hist / total_pixels
b_hist_prob = b_hist / total_pixels
gray_hist_prob = gray_hist / total_pixels

plt.figure(figsize=(15, 10))

# Grayscale Histogram for M1
plt.subplot(2, 2, 1)
plt.plot(gray_hist, color='gray')
plt.title('Grayscale Image Histogram (M1)')
plt.xlabel('Pixel Intensity (Grey Level)')
plt.ylabel('Frequency (Number of Pixels)')

# Grayscale Histogram for M2
plt.subplot(2, 2, 2)
plt.plot(gray_hist_prob, color='gray')
plt.title('Grayscale Image Histogram (M2)')
plt.xlabel('Pixel Intensity (Grey Level)')
plt.ylabel('Probability of Occurrence')

# Color Image Histograms for M1
plt.subplot(2, 2, 3)
plt.plot(r_hist, color='red', label='Red Channel')
plt.plot(g_hist, color='green', label='Green Channel')
plt.plot(b_hist, color='blue', label='Blue Channel')
plt.title('Color Image Histogram (M1)')
plt.xlabel('Pixel Intensity (Grey Level)')
plt.ylabel('Frequency (Number of Pixels)')
plt.legend()

# Color Image Histograms for M2
plt.subplot(2, 2, 4)
plt.plot(r_hist_prob, color='red', label='Red Channel')
plt.plot(g_hist_prob, color='green', label='Green Channel')
plt.plot(b_hist_prob, color='blue', label='Blue Channel')
plt.title('Color Image Histogram (M2)')
plt.xlabel('Pixel Intensity (Grey Level)')
plt.ylabel('Probability of Occurrence')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Q3.
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg', cv2.IMREAD_GRAYSCALE)

# Increase brightness
bright_image = cv2.add(image, 50)

# Decrease brightness
dark_image = cv2.subtract(image, 50)

# histograms
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_bright = cv2.calcHist([bright_image], [0], None, [256], [0, 256])
hist_dark = cv2.calcHist([dark_image], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")

plt.subplot(2, 3, 2)
plt.imshow(bright_image, cmap='gray')
plt.title("Brighter Image")

plt.subplot(2, 3, 3)
plt.imshow(dark_image, cmap='gray')
plt.title("Darker Image")

plt.subplot(2, 3, 4)
plt.plot(hist_original, color='black')
plt.title("Original Histogram")

plt.subplot(2, 3, 5)
plt.plot(hist_bright, color='red')
plt.title("Bright Histogram")

plt.subplot(2, 3, 6)
plt.plot(hist_dark, color='blue')
plt.title("Dark Histogram")

plt.tight_layout()
plt.show()
```

```python
#Q4.
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg', cv2.IMREAD_GRAYSCALE)

# Histogram Equalization
equalized_image = cv2.equalizeHist(image)

hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Image")

plt.subplot(2, 2, 3)
plt.plot(hist_original, color='black')
plt.title("Original Histogram")

plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='red')
plt.title("Equalized Histogram")

plt.tight_layout()
plt.show()
```

```python
#TASK 2 Q1.
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'grayscale_image.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Compute the FFT
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')

plt.show()
```

```python
#Q2.
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'grayscale_image.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# FFT
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)

# Inverse FFT
f_ishift = np.fft.ifftshift(fshift)
img_reconstructed = np.fft.ifft2(f_ishift)
img_reconstructed = np.abs(img_reconstructed)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Reconstructed Image")
plt.imshow(img_reconstructed, cmap='gray')
plt.axis('off')

plt.show()
```

```python
#Q3.
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import rotate

image_path = 'grayscale_image.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Rotate - 45 deg
rotated_image = rotate(image, 45, reshape=True)

# Compute the FFT of original and rotated images
f_orig = np.fft.fft2(image)
fshift_orig = np.fft.fftshift(f_orig)
magnitude_spectrum_orig = 20 * np.log(np.abs(fshift_orig) + 1)

f_rot = np.fft.fft2(rotated_image)
fshift_rot = np.fft.fftshift(f_rot)
magnitude_spectrum_rot = 20 * np.log(np.abs(fshift_rot) + 1)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Rotated Image (45)")
plt.imshow(rotated_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Magnitude Spectrum of Original")
plt.imshow(magnitude_spectrum_orig, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Magnitude Spectrum of Rotated")
plt.imshow(magnitude_spectrum_rot, cmap='gray')
plt.axis('off')

plt.show()
```
