```python
from PIL import Image
import os

image_path = "Dog.jpg"
original_image = Image.open(image_path)

# Perform lossy compression using JPEG
jpeg_path = "compressed_image.jpeg"
original_image.save(jpeg_path, "JPEG", quality=30)
print(f"Lossy compressed image saved as {jpeg_path}")

# Perform lossless compression using PNG
png_path = "compressed_image.png"
original_image.save(png_path, "PNG", optimize=True)
print(f"Lossless compressed image saved as {png_path}")

from IPython.display import display
print("Original Image:")
display(original_image)
print("Lossy Compressed (JPEG) Image:")
display(Image.open(jpeg_path))
print("Lossless Compressed (PNG) Image:")
display(Image.open(png_path))

# Show file sizes
original_size = os.path.getsize(image_path) / 1024  # KB
jpeg_size = os.path.getsize(jpeg_path) / 1024  # KB
png_size = os.path.getsize(png_path) / 1024  # KB

print(f"Original Image Size: {original_size:.2f} KB")
print(f"JPEG Compressed Image Size: {jpeg_size:.2f} KB")
print(f"PNG Compressed Image Size: {png_size:.2f} KB")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist, cifar10
from keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt

# Function to build a simple CNN model
def build_cnn(input_shape, num_classes):
    model = keras.Sequential([
        keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Conv2D(64, (3, 3), activation='relu'),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Conv2D(128, (3, 3), activation='relu'),
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Function to evaluate the model
def evaluate_model(model, X_test, y_test):
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_test_labels = np.argmax(y_test, axis=1)

    accuracy = np.mean(y_pred == y_test_labels)
    class_report = classification_report(y_test_labels, y_pred, output_dict=True)
    conf_matrix = confusion_matrix(y_test_labels, y_pred)

    return accuracy, class_report, conf_matrix

def plot_roc_auc(model, X_test, y_test, num_classes):
    y_prob = model.predict(X_test)
    fpr, tpr, roc_auc = {}, {}, {}
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    plt.figure()
    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()

# Training and Evaluating on MNIST
def train_mnist():
    (X_train, y_train), (X_test, y_test) = mnist.load_data()
    X_train, X_test = X_train / 255.0, X_test / 255.0
    X_train = X_train.reshape(-1, 28, 28, 1)
    X_test = X_test.reshape(-1, 28, 28, 1)
    y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)

    model = build_cnn((28, 28, 1), 10)
    model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=64)

    accuracy, class_report, conf_matrix = evaluate_model(model, X_test, y_test)
    plot_roc_auc(model, X_test, y_test, 10)

    return accuracy, class_report, conf_matrix

# Training and Evaluating on CIFAR-10
def train_cifar10():
    (X_train, y_train), (X_test, y_test) = cifar10.load_data()
    X_train, X_test = X_train / 255.0, X_test / 255.0
    y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)

    model = build_cnn((32, 32, 3), 10)
    model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=64)

    accuracy, class_report, conf_matrix = evaluate_model(model, X_test, y_test)
    plot_roc_auc(model, X_test, y_test, 10)

    return accuracy, class_report, conf_matrix

mnist_results = train_mnist()
cifar10_results = train_cifar10()

print("MNIST Results:", mnist_results)
print("CIFAR-10 Results:", cifar10_results)
```
