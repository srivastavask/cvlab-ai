```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(title, image, cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('data/Picture1.jpg')
color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Task 1: Histogram Analysis and Contrast Enhancement

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display_image("Grayscale Image", gray_image, cmap='gray')
```

```python
def plot_histogram(image, title, color):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.title(title)
    plt.xlabel('Gray Level / Intensity')
    plt.ylabel('Pixel Count')
    plt.show()
```

```python
plot_histogram(gray_image, "Histogram for Grayscale Image", 'black')
```

```python
channels = cv2.split(image)
colors = ('blue', 'green', 'red')

for (channel, color) in zip(channels, colors):
    plot_histogram(channel, f"Histogram for {color} Channel", color)
```

```python
hist_norm = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
hist_norm = hist_norm / hist_norm.sum()
plt.plot(hist_norm, color='gray')
plt.title("Normalized Histogram for Grayscale Image (Probability)")
plt.xlabel('Gray Level / Intensity')
plt.ylabel('Probability')
plt.show()
```

```python
equalized_image = cv2.equalizeHist(gray_image)
display_image("Equalized Grayscale Image", equalized_image, cmap='gray')
```
