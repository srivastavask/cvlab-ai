```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_PATH = 'pic.jpeg'
```

```python
def show_img(header, arr, m=None):
    plt.figure(figsize=(6, 6))
    plt.imshow(arr, cmap=m)
    plt.title(header)
    plt.axis('off')
    plt.show()
```

```python
img_original = cv2.imread(IMG_PATH)
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
```

## Task 1: Histogram Analysis and Contrast Enhancement

```python
img_gray = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
show_img('Grayscale Image', img_gray, m='gray')
```

```python
def render_hist(data, head, shade):
    hvals = cv2.calcHist([data], [0], None, [256], [0, 256])
    plt.plot(hvals, color=shade)
    plt.title(head)
    plt.xlabel('Gray Level / Intensity')
    plt.ylabel('Pixel Count')
    plt.show()
```

```python
render_hist(img_gray, 'Histogram for Grayscale Image', 'black')
```

```python
bgr_planes = cv2.split(img_original)
bgr_names = ['blue', 'green', 'red']
for plane, nm in zip(bgr_planes, bgr_names):
    render_hist(plane, f'Histogram for {nm} Channel', nm)
```

```python
hist_p = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
hist_p = hist_p / np.sum(hist_p)
plt.plot(hist_p, color='gray')
plt.title('Normalized Histogram for Grayscale Image (Probability)')
plt.xlabel('Gray Level / Intensity')
plt.ylabel('Probability')
plt.show()
```

```python
img_eq = cv2.equalizeHist(img_gray)
show_img('Equalized Grayscale Image', img_eq, m='gray')
```
